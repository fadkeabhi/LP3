## UsIng Relevant Features

```python
data = df[['SALES', 'QUANTITYORDERED', 'PRICEEACH']].copy()
```

```python
data
```

```python
data.isnull().sum()
```

## Scaling the Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
data_scaled
```

```python
sse = []
k_range = range(1, 11)
```

## Detemining the no. of clusters using Elbow Method

```python
from sklearn.cluster import KMeans
```

```python
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    sse.append(kmeans.inertia_)
```

```python
sse
```

```python
plt.figure(figsize=(8, 5))
plt.plot(k_range, sse, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.show()
```

## Training the Model

```python
optimal_k = 3  
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(data_scaled)
```

```python
df['Cluster'] = kmeans.labels_
```

```python
df
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=data_scaled[:, 0], y=data_scaled[:, 1], hue=kmeans.labels_, palette='viridis')
plt.title('Clusters of Sales Data')
plt.xlabel('Scaled Feature 1 (SALES)')
plt.ylabel('Scaled Feature 2 (QUANTITYORDERED)')
plt.show()
```
