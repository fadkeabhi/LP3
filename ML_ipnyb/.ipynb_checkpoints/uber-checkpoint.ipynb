```python
df = df[(df["dropoff_longitude"] >= -180) & (df["dropoff_longitude"] <= 180)]
df = df[(df["dropoff_latitude"] >= -90) & (df["dropoff_latitude"] <= 90)]
```

```python
df.reset_index(inplace=True)
```

```python
pip install haversine
```

```python
import haversine as hv
```

```python
def dist(row):
    l1 = (row["pickup_latitude"], row["pickup_longitude"])
    l2 = (row["dropoff_latitude"], row["dropoff_longitude"])
    return hv.haversine(l1, l2)
```

```python
df["distance"] = df.apply(dist, axis=1)
```

```python
import numpy as np
```

```python
q1 = df["passenger_count"].quantile(0.25)
q3 = df["passenger_count"].quantile(0.75)
iqr = q3-q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
df["passenger_count"] = np.clip(df["passenger_count"], lower, upper)
```

```python
plt.boxplot(x=df["passenger_count"])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df.reset_index()
```

```python
df.drop(['pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude'], inplace=True, axis=1)
```

```python
df.head()
```

```python
x = df.drop("fare_amount", axis=1)
y = df["fare_amount"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2_score(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
rmodel = RandomForestRegressor()
```

```python
rmodel.fit(x_train, y_train)
```

```python
y_pred = rmodel.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```
